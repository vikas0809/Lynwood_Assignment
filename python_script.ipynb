```python
import pandas as pd
```

```python
xls = pd.ExcelFile("Data/Data Analyst Assignment.xlsx")
cti_df = pd.read_excel(xls, 'CTI Report')
feedback_df = pd.read_excel(xls, 'Feedback Report')
```

```python
cti_df.head()
```

```python
cti_df.isna().sum()
```

```python
cti_df.dropna(inplace=True)
```

```python
cti_df.isna().sum()
```

```python
duplicate_df=cti_df[cti_df["Client ID"].duplicated(keep=False)]
duplicate_df=duplicate_df.sort_values(by='Client ID')
duplicate_df.head()
```

```python
index_list=duplicate_df.index
index_list
```

```python
index_to_drop=[]
for index in index_list:
    cl_id= cti_df['Client ID'].loc[index]
    for index1 in index_list:
        if (index != index1) and (cl_id == cti_df['Client ID'].loc[index1]):
            if cti_df['CTI completion date'].loc[index] >= cti_df['CTI completion date'].loc[index1]:
                index_to_drop.append(index1)

index_to_drop
```

```python
clean_cti_df=cti_df.drop(index=index_to_drop)
clean_cti_df.head(10)
```

```python
clean_cti_df[clean_cti_df["Client ID"].duplicated(keep=False)]
```

```python
clean_cti_df['Age Category'] = ["Youth" if x >= 12 else "Child" for x in clean_cti_df['Age at Assessment']]
clean_cti_df.head(10)
```

```python
feedback_df.isna().sum()
```

```python
duplicate_df=feedback_df[feedback_df["Client ID"].duplicated(keep=False)]
duplicate_df=duplicate_df.sort_values(by='Client ID')
duplicate_df.head()
```

```python
duplicate_df['Client ID'].nunique()
```

```python
index_list=duplicate_df.index
index_list
```

```python
index_to_drop=[]
for index in index_list:
    cl_id= feedback_df['Client ID'].loc[index]
    for index1 in index_list:
        if (index != index1) and (cl_id == feedback_df['Client ID'].loc[index1]):
            if feedback_df['Completion Date'].loc[index] >= feedback_df['Completion Date'].loc[index1]:
                index_to_drop.append(index1)

index_to_drop
```

```python
clean_feedback_df=feedback_df.drop(index=index_to_drop)
clean_feedback_df.head(10)
```

```python
clean_feedback_df[clean_feedback_df["Client ID"].duplicated(keep=False)]
```

```python
merged_df=clean_cti_df.merge(clean_feedback_df,how='inner',on='Client ID')
merged_df.head()
```

```python
merged_df.to_excel("Data/merged_report.xlsx")
```
